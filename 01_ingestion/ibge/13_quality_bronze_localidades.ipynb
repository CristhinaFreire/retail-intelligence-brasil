```python
resultado = []

for tabela in TABLES:

    print("=" * 80)
    print(f"Tabela: {tabela}")

    full_table = f"{CATALOG}.{SCHEMA}.{tabela}"

    try:

        df = spark.table(full_table)

        registros = df.count()
        colunas = len(df.columns)

        resultado.append({
            "Tabela": tabela,
            "Registros": registros,
            "Colunas": colunas,
            "Status": "OK"
        })

        print(f"✓ Registros : {registros}")
        print(f"✓ Colunas   : {colunas}")

        df.printSchema()

        display(df.limit(5))

    except Exception as e:

        resultado.append({
            "Tabela": tabela,
            "Registros": 0,
            "Colunas": 0,
            "Status": "ERRO"
        })

        print(e)
```

```python
quality_df = (
    spark
    .createDataFrame(resultado)
)

display(quality_df)
```
