```python
# ============================================================
# Retail Intelligence Brasil
# Notebook.......: 13_quality_bronze_localidades
# Camada.........: Quality
# Fonte..........: Delta Lake - Bronze Localidades
# Objetivo.......: Validar a disponibilidade estrutural das tabelas Bronze de localidades.
# Autor..........: Cristhina Freire
# Data...........: 26/06/2026
# ============================================================

# ============================================================
# 1. CONFIGURAÇÃO
# ============================================================

CATALOG = "retail_intelligence"
SCHEMA = "bronze"

TABLES = [
    "ibge_estados_raw",
    "ibge_municipios_raw",
    "ibge_regioes_raw",
    "ibge_regioes_intermediarias_raw",
    "ibge_regioes_imediatas_raw",
    "ibge_mesorregioes_raw",
    "ibge_microrregioes_raw",
    "ibge_distritos_raw",
    "ibge_subdistritos_raw",
    "ibge_regioes_metropolitanas_raw",
    "ibge_aglomeracoes_urbanas_raw",
]

print("=" * 70)
print("13_QUALITY_BRONZE_LOCALIDADES")
print("=" * 70)
```

```python
# ============================================================
# 2. INSPEÇÃO
# ============================================================

resultado = []

for tabela in TABLES:

    print("=" * 70)
    print(f"Tabela: {tabela}")

    FULL_TABLE_NAME = f"{CATALOG}.{SCHEMA}.{tabela}"

    try:
        df = spark.table(FULL_TABLE_NAME)

        registros = df.count()
        colunas = len(df.columns)

        resultado.append({
            "Tabela": tabela,
            "Registros": registros,
            "Colunas": colunas,
            "Status": "OK"
        })

        print(f"Registros : {registros}")
        print(f"Colunas   : {colunas}")

        df.printSchema()
        display(df.limit(5))

    except Exception as error:
        resultado.append({
            "Tabela": tabela,
            "Registros": 0,
            "Colunas": 0,
            "Status": "ERRO"
        })

        print(error)
```

```python
# ============================================================
# 3. CONSOLIDAÇÃO
# ============================================================

quality_df = (
    spark
    .createDataFrame(resultado)
)

display(quality_df)

print("=" * 70)
print("Pipeline finalizado com sucesso.")
print("=" * 70)
```
