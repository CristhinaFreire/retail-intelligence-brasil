```python
# ============================================================
# 1. IMPORTS
# ============================================================

import requests
import json
from datetime import datetime

```

```python
# ============================================================
# 2. CONFIGURAÇÃO
# ============================================================

BASE_URL = "https://servicodados.ibge.gov.br/api/v1/localidades/estados"

TIMEOUT = 30

CURRENT_DATE = datetime.now().strftime("%Y-%m-%d")
CURRENT_TIME = datetime.now().strftime("%Y%m%d_%H%M%S")

VOLUME_PATH = (
    f"/Volumes/retail_intelligence/bronze/landing/ibge/"
    f"estados/{CURRENT_DATE}"
)

FILE_NAME = f"estados_{CURRENT_TIME}.json"

print("=" * 60)
print("DOWNLOAD - ESTADOS (IBGE)")
print("=" * 60)

```

```python
# ============================================================
# 3. EXTRAÇÃO
# ============================================================

print("Consumindo API do IBGE...")

response = requests.get(
    url=BASE_URL,
    timeout=TIMEOUT
)

```

```python
# ============================================================
# 4. VALIDAÇÃO
# ============================================================

if response.status_code != 200:
    raise Exception(
        f"Erro ao consumir API. Status Code: {response.status_code}"
    )

landing_data = response.json()

print("API consumida com sucesso.")
print(f"Total de registros: {len(landing_data)}")

```

```python
# ============================================================
# 5. PERSISTÊNCIA LANDING
# ============================================================

dbutils.fs.mkdirs(VOLUME_PATH)

FILE_PATH = f"{VOLUME_PATH}/{FILE_NAME}"

dbutils.fs.put(
    FILE_PATH,
    json.dumps(
        landing_data,
        ensure_ascii=False,
        indent=2
    ),
    overwrite=True
)

print(f"\nJSON salvo em:")
print(FILE_PATH)

```

```python
# ============================================================
# 6. CONVERTER PARA SPARK
# ============================================================

landing_df = spark.createDataFrame(landing_data)

print("\nSchema:")

landing_df.printSchema()

print("\nPrévia dos dados:")

display(landing_df)

```
