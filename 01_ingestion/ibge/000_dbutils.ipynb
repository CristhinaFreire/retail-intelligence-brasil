```python
# ============================================================
# Retail Intelligence Brasil
# Notebook.......: 000_dbutils
# Camada.........: Utilities
# Fonte..........: Databricks Volumes / IBGE Landing
# Objetivo.......: Reorganizar os diretórios de landing de localidades do IBGE.
# Autor..........: Cristhina Freire
# Data...........: 25/06/2026
# ============================================================

# ============================================================
# 1. CONFIGURAÇÃO
# ============================================================

BASE_PATH = "/Volumes/retail_intelligence/bronze/landing/ibge"

print("=" * 70)
print("000_DBUTILS")
print("=" * 70)
```

```python
# ============================================================
# 2. REORGANIZAÇÃO DAS PASTAS
# ============================================================

dbutils.fs.mv(f"{BASE_PATH}/estados", f"{BASE_PATH}/localidades/estados")
dbutils.fs.mv(f"{BASE_PATH}/municipios", f"{BASE_PATH}/localidades/municipios")
dbutils.fs.mv(f"{BASE_PATH}/regioes", f"{BASE_PATH}/localidades/regioes")
dbutils.fs.mv(f"{BASE_PATH}/regioes_intermediarias", f"{BASE_PATH}/localidades/regioes_intermediarias")
dbutils.fs.mv(f"{BASE_PATH}/regioes_imediatas", f"{BASE_PATH}/localidades/regioes_imediatas")
dbutils.fs.mv(f"{BASE_PATH}/mesorregioes", f"{BASE_PATH}/localidades/mesorregioes")
dbutils.fs.mv(f"{BASE_PATH}/microrregioes", f"{BASE_PATH}/localidades/microrregioes")
dbutils.fs.mv(f"{BASE_PATH}/distritos", f"{BASE_PATH}/localidades/distritos")
dbutils.fs.mv(f"{BASE_PATH}/subdistritos", f"{BASE_PATH}/localidades/subdistritos")
dbutils.fs.mv(f"{BASE_PATH}/regioes_metropolitanas", f"{BASE_PATH}/localidades/regioes_metropolitanas")
```

```python
# ============================================================
# 3. GARANTIA DA PASTA-BASE
# ============================================================

dbutils.fs.mkdirs("/Volumes/retail_intelligence/bronze/landing/ibge/localidades")
```

```python
# ============================================================
# 4. MIGRAÇÃO EM LOTE
# ============================================================

dbutils.fs.mkdirs(f"{BASE_PATH}/localidades")

pastas = [
    "estados",
    "municipios",
    "regioes",
    "regioes_intermediarias",
    "regioes_imediatas",
    "mesorregioes",
    "microrregioes",
    "distritos",
    "subdistritos",
    "regioes_metropolitanas"
]

for pasta in pastas:
    origem = f"{BASE_PATH}/{pasta}"
    destino = f"{BASE_PATH}/localidades/{pasta}"

    print(f"Movendo {pasta}...")
    dbutils.fs.mv(origem, destino, True)

print("Concluído.")
```

```python
# ============================================================
# 5. INSPEÇÃO
# ============================================================

print("IBGE")
display(dbutils.fs.ls("/Volumes/retail_intelligence/bronze/landing/ibge"))

print("LOCALIDADES")
display(dbutils.fs.ls("/Volumes/retail_intelligence/bronze/landing/ibge/localidades"))

print("=" * 70)
print("Pipeline finalizado com sucesso.")
print("=" * 70)
```
