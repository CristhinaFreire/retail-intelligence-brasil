```python
# ============================================================
# 1. IMPORTS
# ============================================================

import requests
import json
```

```python
# ============================================================
# 2. CONFIGURAÇÃO
# ============================================================

BASE_URL = "https://apisidra.ibge.gov.br"

TABLE_ID = 9514

TIMEOUT = 60

print("=" * 60)
print("SPIKE - CLASSIFICAÇÕES SIDRA")
print("=" * 60)
```

```python
# ============================================================
# 3. METADADOS DA TABELA
# ============================================================

url = f"{BASE_URL}/desctabapi.aspx?c={TABLE_ID}"

print(url)

response = requests.get(url, timeout=TIMEOUT)

print(response.status_code)

print(response.text[:3000])
```
